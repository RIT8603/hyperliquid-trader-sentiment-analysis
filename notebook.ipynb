# Merge Datasets

### # Step 4: Merge Historical Data with Fear & Greed Index

The two datasets are merged using the common `date` column so that each trade is associated with the corresponding market sentiment.

```python
print(historical_data.info())

print(fear_greed.info())
```

```python
fear_greed["date"] = pd.to_datetime(fear_greed["date"])
merged_data = historical_data.merge(
    fear_greed[["date", "classification", "value"]],
    on="date",
    how="left"
)

print(merged_data.shape)
print("merged successful!! ")
```

# EDA


#### Step 5: Exploratory Data Analysis (EDA)

This section explores the trading dataset and identifies relationships between trader performance and market sentiment.

```python
print("Merged Dataset Shape:", merged_data.shape)

merged_data.head()
```

```python
merged_data.isnull().sum()
```

```python
print("Total Traders : ", len(merged_data['Account'].unique()))
```

```python
merged_data['Coin'].value_counts()
```

```python
plt.figure(figsize=(10, 5))

merged_data['Coin'].value_counts().head(10).plot(kind='bar')


plt.title("Top 10 Most Traded Coins")
plt.xlabel("Coin")
plt.ylabel("Frequency")

plt.savefig("Top 10 Most Traded coins.jpg")
plt.show()
```

```python
merged_data['Side'].value_counts()
```

```python
plt.figure(figsize=(6,4))

sns.countplot(data=merged_data,x="Side")

plt.title("Buy vs Sell Trades")
plt.xlabel("Side")
plt.ylabel("Frequency")

plt.savefig("Buy vs Sell Trades.jpg")
plt.show()
```

```python
merged_data["classification"].value_counts()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(data=merged_data,x="classification")
plt.xticks(rotation=20)

plt.title("Fear & Greed Classification")
plt.xlabel("Classification")
plt.ylabel("Frequency")

plt.savefig("Fear & Greed Classification.jpg")
plt.show()
```

```python
# profit distribution

plt.figure(figsize=(10,5))

sns.histplot(merged_data["Closed PnL"],bins=80)
plt.xlabel("Profit")
plt.ylabel("Frequency")
plt.title("Profit / Loss Distribution")

plt.savefig("Profit and Loss Distribution.jpg")
plt.show()
```

```python
# Profit by Classification

plt.figure(figsize=(8, 5))

sns.boxplot(data=merged_data,
            x='classification',
            y='Closed PnL')

plt.xticks(rotation=20)
plt.xlabel("Classification")
plt.ylabel("Profit")
plt.title("Profit by Classification")

plt.savefig("Profit by Classification.jpg")
plt.show()
```

```python
#Average Profit

merged_data.groupby("classification")['Closed PnL'].mean()
```

```python
# Averae Trading Volume
merged_data.groupby("classification")["Size USD"].mean()
```

```python
# Top 10 Profitable Traders
top = merged_data.groupby("Account")["Closed PnL"].sum()

top.sort_values(ascending=False).head(10)
```

```python
top.sort_values(ascending=False).head(10).plot(kind="bar")


plt.title("Top 10 Profitable Traders")
plt.xlabel("Account")
plt.ylabel("Profit")

plt.savefig("Top 10 Profitable Traders.jpg")
plt.show()
```

```python
# Correlation Heatmap

plt.figure(figsize=(10, 8))

sns.heatmap(
    merged_data.corr(numeric_only=True),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")

plt.savefig("Correlation Heatmap.jpg")
plt.show()
```

```python
# Daliy Profit

daily_profit = merged_data.groupby("date")['Closed PnL'].sum()
daily_profit.plot(figsize=(12, 5))

plt.title("Daily Profit")
plt.xlabel("Date")
plt.ylabel("Profit")

plt.savefig("Daily Profit.jpg")
plt.show()
```
